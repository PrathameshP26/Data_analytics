### 1) Scatter Plot (Numerical-Numerical)

```python
plt.scatter(tips['total_bill'],tips['tip'],color='blue',alpha=0.5)
plt.show()
```

```python
sns.scatterplot(x=tips['total_bill'],y=tips['tip'],hue=tips['sex'])
plt.title('Multivariate analysis')
```

```python
sns.scatterplot(x=tips['total_bill'],y=tips['tip'],hue=tips['sex'],style=tips['smoker'])
plt.title('Multivariate analysis')
```

```python

sns.scatterplot(x=tips['total_bill'],y=tips['tip'],hue=tips['sex'],style=tips['smoker'],size=tips['size'])
plt.title('Multivariate analysis')
```

### Bar Plot (Numerical-Categorical)

```python
titanic=sns.load_dataset('titanic')
```

```python
titanic.head(5)
```

```python
plt.figure(figsize=(10, 9))
sns.barplot(data=titanic,x=titanic['pclass'],y=titanic['age'])
```

```python
plt.figure(figsize=(10, 9))
sns.barplot(data=titanic,x=titanic['pclass'],y=titanic['fare'])
```

```python
plt.figure(figsize=(9, 7))
sns.barplot(data=titanic,x=titanic['pclass'],y=titanic['fare'],hue=titanic['sex'])
```

### Box Plot (Numerical-Categorical)

```python
sns.boxplot(data=titanic,x=titanic['sex'],y=titanic['age'],hue=titanic['survived'])
```

### DistPlot Numerical-Categorical

```python
sns.distplot(titanic[titanic['survived']==0]['age'],hist=False)
sns.distplot(titanic[titanic['survived']==1]['age'],hist=False)
plt.legend()
```

### Heat map (categorical-categorical)

```python
sns.heatmap(pd.crosstab(titanic['pclass'],titanic['survived']))
```

```python
titanic.head()
```

### Cluster Map (Categorical-Categorical)

```python
sns.clustermap(pd.crosstab(titanic['sibsp'],titanic['survived']))
```
