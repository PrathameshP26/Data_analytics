## Box Plot

To Detect the Outliers we can use Box plot

IQR = Inter Quartile Range

Q1 = 25th percentile

Q3 = 75th percentile

Minimum = Q1-1.5*IQR

Maximum = Q3+1.5*IQR


There are 3 types of skewness

1) Symetric Skewed

2) Positive Skewed

3) Negatively Skewed
