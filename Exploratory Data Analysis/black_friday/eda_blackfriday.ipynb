merge both train and test data

```python
df.drop(['User_ID'],axis=1,inplace=True)
```

Handling categorical feature 'Gender'

```python
df['Gender'] = df['Gender'].map({'F':0,'M':1})
'''
or 

df['Gender']=pd.get_dummies(df['Gender'],drop_first=True)
'''
df.head()
```

Handling categorical feature age

```python
## Handling categorical feature age

df['Age'].unique()
```

```python
# we can perform ordinal encoding
df['Age']=df['Age'].map({'0-17':1,'18-25':2,'26-35':3,'36-45':4,'46-50':5,'51-55':6,'55+':7})
```

```python
df.head()
```

Handling categorical feature 'city' 

```python
city=pd.get_dummies(df['City_Category'],drop_first=1)
```

```python
df=pd.concat([df,city],axis=1)
df.head()
```

```python
df.drop(['City_Category'],inplace=True,axis=1)
```

```python
df.head()
```

Handling the missing values

```python
df.isnull().sum()
```

```python
df['Product_Category_2'].value_counts()
```

```python
# here the product category 2 is discrete values so we replace nan with mode

mode_prod_2=df['Product_Category_2'].mode()[0] # why 0 ? because its a series
```

```python
df['Product_Category_2']=df['Product_Category_2'].fillna(mode_prod_2)
```

```python
df.isnull().sum()
```
